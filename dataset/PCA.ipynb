```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

from numpy import genfromtxt
params = genfromtxt('file.csv', delimiter=',')
```

```python
for student in range(12):
    
    for cut in range(12):
        y_hat = np.zeros(180)
        for i in range(20):
            y_hat[i*9:i*9+9] = params[student*12+cut,i]
        plt.plot(y_hat,alpha=0.4)

    mu = np.mean(params[student*12:student*12+12],axis=0)
    sd = np.std(params[student*12:student*12+12],axis=0)
    y_mu = np.zeros(180)
    y_sdu = np.zeros(180)
    y_sdl = np.zeros(180)
    for i in range(20):
        y_mu[i*9:i*9+9] = mu[i]
        y_sdu[i*9:i*9+9] = y_mu[i*9:i*9+9] + sd[i]
        y_sdl[i*9:i*9+9] = y_mu[i*9:i*9+9] - sd[i]
    plt.plot(y_mu,'--',c='k'); plt.plot(y_sdu,':',c='k',alpha=0.7); plt.plot(y_sdl,':',c='k',alpha=0.7);
    
    plt.title('Student '+str(student+1))
    plt.ylim([-2,2])
    plt.show()
```

```python
X_standardized = StandardScaler().fit_transform(params)
print(X_standardized.shape)
print(np.mean(X_standardized,axis=0))
print(np.std(X_standardized,axis=0))
```

```python
def plot_embedding(X, y_, title=None):
  x_min, x_max = np.min(X, 0), np.max(X, 0)
  X = (X - x_min) / (x_max - x_min)

  plt.figure(figsize=(10,10))
  ax = plt.subplot(111)
  for i in range(X.shape[0]):
    plt.text(X[i, 0], X[i, 1], str(y_[i]),
              color=plt.cm.Set1(y_[i] / 11.),
              fontdict={'weight': 'bold', 'size': 12})

  plt.xticks([]), plt.yticks([])
  if title is not None:
    plt.title(title)
```

```python
y = np.ones(params.shape[0],dtype='int32')
for i in range(12):
    y[i*12:i*12+12] = i+1
    
print(y)
```

```python
# PCA
from sklearn import decomposition

X_pca = decomposition.TruncatedSVD(n_components=2).fit_transform(X_standardized)
plot_embedding(X_pca,y,'PCA')
```

```python
# LDA
from sklearn import discriminant_analysis

X2 = X_standardized.copy()
# X2.flat[::X_standardized.shape[1] + 1] += 0.01  # Make X invertible
X_lda = discriminant_analysis.LinearDiscriminantAnalysis(n_components=2).fit_transform(X_standardized, y)
plot_embedding(X_lda,y,'LDA')
```

```python
from sklearn import manifold

tsne = manifold.TSNE(n_components=2, init='pca', random_state=0)

X_tsne = tsne.fit_transform(X_standardized)

plot_embedding(X_tsne,y,'T-SNE')
```

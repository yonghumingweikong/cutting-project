```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mass_spring_damper(x,u,dt):
    
    k = 2
    b = 4
    m = 1.5
    
    A = np.array([[0.0,1.0],[-k/m,-b/m]])
    B = np.array([0.0,1.0/m])
    
    x_new = x + np.matmul(A,x)*dt + B*u*dt
    
    return x_new
```

```python
# Open loop
N = 180
dt = 0.0333
t = np.arange(0,N*dt,dt)

x0 = np.zeros(2)
x = [x0]

for i in range(t.shape[0]-1):
    x.append(mass_spring_damper(x[-1],2,dt))

plt.plot(t,np.array(x)); plt.grid()
```

```python
# Close loop
N = 180
dt = 0.0333
t = np.arange(0,N*dt,dt)

x0 = np.zeros(2)
xT = 1
Kp = 20
Ki = 1
Kd = 10

x = [x0]
err = [0]
u = [0]

for i in range(t.shape[0]):
    err.append(xT - x[-1][0])
    u.append(Kp*err[-1] - Kd*x[-1][1] + Ki*np.sum(np.array(err)))
    x.append(mass_spring_damper(x[-1],u[-1],dt))

plt.plot(t,np.array(x)[:-1]); plt.grid()
```

```python
plt.plot(t,np.array(u)[:-1])
plt.grid()
```

```python
def double_integrator(state,u,dt):
    A = np.array([[0,1],[0,0]])
    B = np.array([0,1])
    new_state = state + np.matmul(A,state)*dt + B*u*dt
    return new_state
```

```python
N = 180
dt = 0.0333
t = np.arange(0,N*dt,dt)

acc = [0.1,0.5,1.0]

for j in range(3):
    states = [np.zeros(2)]
    controls = np.zeros(t.shape[0])
    controls[0:100] = acc[j]

    for i in range(t.shape[0]):
        err = xT - states[-1][0]
    #     controls.append(Kp*err-Kd*states[-1][1])
        states.append(double_integrator(states[-1],controls[i],dt))

    plt.plot(np.array(states)[:,0])

    plt.grid()
```

```python
# Generate the reference velocity profile
N = 180
dt = 0.0333
t = np.arange(0,N*dt,dt)

xT = 30
states = [np.zeros(2)]
controls = np.zeros(t.shape[0])
controls[0:100] = 1

vref = [0]

for i in range(t.shape[0]):
    err = xT - states[-1][0]
    states.append(double_integrator(states[-1],controls[i],dt))
    vref.append(states[-1][1])

plt.plot(np.array(states))
plt.grid(); plt.show()

plt.plot(np.array(vref)); plt.grid()
```

```python
N = 180
dt = 0.0333
t = np.arange(0,N*dt,dt)

Kp = 10
Kd = 1

states = [np.zeros(2)]
controls = [0]
pert = np.zeros(t.shape[0]); # perturbation
pert[100:115] = 5; pert[115:150] = 0; 

for i in range(t.shape[0]):
    err = vref[i]+pert[i] - states[-1][1]
    u = Kp*err -Kd*states[-1][1]
    controls.append(u);
    states.append(double_integrator(states[-1],u,dt))

plt.plot(np.array(states)[:,1])
plt.plot(vref,'--')
plt.grid(); plt.show()

plt.plot(np.array(states)[:,0]); plt.grid(); plt.show();

plt.plot(np.array(controls)); plt.grid()
```

```python
def strain_rate2force(state,u,dt):
    return state -4*state*dt + u*dt
```

```python
import numpy as np
from matplotlib import pyplot as plt
from IPython import display
```

Let $$\mathbf{x}_t = f(\mathbf{x}_{t-1},\mathbf{u}_t)$$, where $\mathbf{x}_t = [x_t,y_t]^\text{T}.$ Assume we have an approximate linear dynamics model, $$\mathbf{x}_t \sim \mathbf{A}_t\mathbf{x}_{t-1}  + \mathbf{B} \mathbf{u}_t$$

Then, assuming a quadratic cost $$J = \sum_{t=0}^T \mathbf{x}_t \mathbf{Q} \mathbf{x}_t^\text{T} + \mathbf{u}_t \mathbf{R} \mathbf{u}_t^\text{T}$$

the optimal control $\mathbf{u}^*$ minimising the cost is given by a linear quadratic regulator.
$$\mathbf{K}_{lqr} =  LQR(\mathbf{A},\mathbf{B},\mathbf{Q},\mathbf{R}) $$
$$\mathbf{u}^* = \mathbf{K}_{lqr}(\mathbf{x}_t-\mathbf{x}_d)$$

where $\mathbf{x}_d$ is some desired state.

Given these assumptions, the inverse optimal control problem can be framed as an inference problem:

Find $\mathbf{Q}^*,\mathbf{R}^*$ such that $\mathbf{u}^* \sim \mathbf{u}^h$, where $\mathbf{u}^h$ is a set of actions demonstrated by a human.

This problem formulation dominates inverse optimal control at present, but makes two important assumptions, which we challenge here.

* Humans are optimal - we strongly dispute this
* The model (A,B) is accurate - problematic because Doyle (LQR controllers are flawed when model-plant mismatch)

What's missing?
* Feedback - Humans have a capacity to adapt to modelling error

Assume human policy is to act optimally according to model (linear quadratic regulator), and use feedback to compensate for model-plant mismatch.
$$\mathbf{u}_t^* = \mathbf{K}_{lqr}(\mathbf{x}_t-\mathbf{x}_d) + \mathbf{K}_{fb}(\mathbf{A}_{t-1}\mathbf{x}_{t-1} + \mathbf{B}_{t-1}\mathbf{u}_{t-1} - \mathbf{x}_t).$$

Given these assumptions, the inverse optimal control problem can still be framed as an inference problem:

Find $\mathbf{Q}^*,\mathbf{R}^*$, $\mathbf{K}_{fb}$ such that $\mathbf{u}^* \sim \mathbf{u}^h$, where $\mathbf{u}^h$ is a set of actions demonstrated by a human, and by estimating the disturbance error term 

$$(\mathbf{A}_{t-1}\mathbf{x}^h_{t-1} + \mathbf{B}_{t-1}\mathbf{u}^h_{t-1} - \mathbf{x}^h_t)$$ using human state $\mathbf{x}^h_t$ and action pairs.





```python
class PointMass:
    def __init__(self, dt=0.1):
        
        self.x = 0.0 # y-position
        self.y = 0.0 # y-position 
        
        self.vx = 0.0 #rotational velocity left wheel
        self.vy = 0.0 #rotational velocity right wheel
        
        self.dt = dt
        self.slip = 0.5*np.random.randn()
        
    def motor_sim(self,v,torque,sim=True):
    
        v = v + self.dt*torque
    
        return v
    
    
    # Kinematic motion model
    def pose_update(self,torque_x,torque_y,sim=True):
        
        if not sim:
            self.slip = self.slip+0.01*np.random.randn()
            self.vx = self.motor_sim(self.vx,torque_x,sim) + self.slip
            self.vy = self.motor_sim(self.vy,torque_y,sim) + self.slip
        else:
            self.vx = self.motor_sim(self.vx,torque_x,sim)
            self.vy = self.motor_sim(self.vy,torque_y,sim)
        
        self.x = self.x + self.dt*self.vx
        self.y = self.y + self.dt*self.vy
        
        return self.x, self.y
```

```python
class RobotController:
    
    def __init__(self,Kp=0.1,Ki=0.01):
        
        self.Kp = Kp
        self.Ki = Ki
        
    def p_control(self,vx_desired,vy_desired,vx,vy):
        
        torque_x = self.Kp*(vx_desired-vx)
        torque_y = self.Kp*(vy_desired-vy)
        
        return torque_x, torque_y
        
        
    def drive(self,vx_desired,vy_desired,vx,vy):
        
        
        torque_x, torque_y = self.p_control(vx_desired,vy_desired,vx,vy)
        
        return torque_x, torque_y
```

```python
class RobotPathFollowingController:
    
    def __init__(self):
        
        self.controller = RobotController(Kp=5,Ki=0.25)
    
    def get_closest_point(self,traj,X):
        dists = np.sqrt(np.sum((traj - X)**2,axis=1))
        idx = np.argmin(dists)
        return idx
    
    def ref_dynamics(self,dt=0.05):
        
        A = np.array([[1,0],[0,1]])
        B = np.array([[dt, 0],[0,dt]])
        
        return A,B
    
    def lqr(self,A,B,S,Q,R):
        K = np.matmul(np.matmul(np.matmul(np.linalg.inv(np.matmul(np.matmul(B.T,S),B) + R),B.T),S),A)
        S = np.matmul(np.matmul(A.T,S),(A-np.matmul(B,K))) + Q
        return np.copy(K), np.copy(S)
```

```python
robot = PointMass(dt=0.15)
path_controller = RobotPathFollowingController()
```

```python
# Set up target trajectory
target_y = np.linspace(-1,1,100)
target_x = 0.2*np.cos(np.pi*target_y) + 0.2
X_traj = np.vstack((target_x[1:],target_y[1:])).T

plt.plot(X_traj[::10,0],X_traj[::10,1],'-x')
plt.grid()
plt.ylim(-2,2)
plt.xlim(-2,2)

plt.show()
```

#### Generate a demonstration (from a plant with slip, feedback control)

```python
plt.figure(figsize=(15,5))

X0 = [0,-1]

robot.x = X0[0]
robot.y = X0[1]
robot.vx = 0
robot.vy = 0


lookahead = 5

states_gt = [X0]
controls_gt = [[0,0]]

Q = np.eye(2)
R = 0.05*np.eye(2)

Kfb = 6*np.eye(2)

Nsteps = 50

dt = 0.15 # Let's assume it's touch stimulus, with an initial visual reference prior
idx = 0

for j in range(Nsteps):
    # Find closest reference point in front of knife
    idx = min(idx+path_controller.get_closest_point(X_traj[idx:,0:2],states_gt[-1][0:2])+lookahead,X_traj.shape[0]-1)

    A,B = path_controller.ref_dynamics(dt=dt)

    # Solve discrete ricatti a couple times
    S = Q
    K,S = path_controller.lqr(A,B,S,Q,R)

    path_err = states_gt[j] - X_traj[idx,:]

    if j>1:
        u = np.matmul(-K,path_err) + Kfb.dot(A.dot(states_gt[j-1]) + B.dot(u)-states_gt[j])
    else:
        u = np.matmul(-K,path_err)
    
    torque_x,torque_y = path_controller.controller.drive(u[0],u[1],robot.vx,robot.vy)

    x,y = robot.pose_update(torque_x,torque_y,sim=False)

    states_gt.append(np.hstack((x,y)))
    controls_gt.append(u)

    plt.clf()
    plt.subplot(2,2,1)
    plt.plot(X_traj[::10,0],X_traj[::10,1],'-x',label='Desired')
    plt.plot(np.array(states_gt)[:,0],np.array(states_gt)[:,1],'-o',label='Controlled')
    plt.legend()
    plt.title(torque_x)
    plt.ylabel('Y-position')
    plt.xlabel('X-position')
    plt.grid()
    plt.xlim(-2,2)
    plt.ylim(-1.2,1.2)

    plt.subplot(2,2,3)
    plt.plot(np.array(controls_gt))
    plt.xlabel('Time (seconds)')
    plt.legend(['Vx','Vy'])
    plt.grid()

    display.clear_output(wait=True)
    display.display(plt.gcf())
```

#### Infer LQR cost from demonstration

```python
def generative_controller_model(robot,controller,X_traj,sigQ=1,sigR=0.01,Kfb = np.ones((2,2)),lookahead=5, Nsteps=50, dt=0.15):
    
    X0 = [0,-1]
    
    robot.x = X0[0]
    robot.y = X0[1]
    
    robot.vx = 0
    robot.vy = 0

    states = [X0]
    u = [0,0]
    
    controls = [u]
    
    Q = sigQ**2*np.eye(2)
    R = sigR**2*np.eye(2)

    idx = 0

    for j in range(Nsteps):
        
        # Find closest reference point in front of knife
        idx = min(idx+path_controller.get_closest_point(X_traj[idx:,0:2],states_gt[j][0:2])+lookahead,X_traj.shape[0]-1)

        A,B = path_controller.ref_dynamics(dt=dt)

        # Solve discrete ricatti a couple times
        S = Q
        K,S = path_controller.lqr(A,B,S,Q,R)

        path_err = states_gt[j] - X_traj[idx,:]
        
        if j > 1:
            u = np.matmul(-K,path_err) + Kfb.dot((A.dot(states_gt[j-1])+B.dot(controls_gt[j])) - states_gt[j])
        else:
            u = np.matmul(-K,path_err)

        controls.append(u)
        
    return controls
```

```python
def cost(X):
    X = np.abs(X)
    controls_gen = generative_controller_model(robot,path_controller,X_traj,sigQ=X[0],sigR=X[1],Kfb=X[2:].reshape(2,2),lookahead=5, Nsteps=50, dt=0.15)
    c = np.sum((np.array(controls_gt)-np.array(controls_gen))**2) #+ np.sum(np.array(states_gt)-np.array(states_gen))**2
    return c
```

```python
from scipy.optimize import minimize, fmin 
```

```python
# c_weights = np.abs(cross_entropy_min(cost,mu_init=np.ones(4),Ns=200,Nreps=50,d=4))
```

```python
c_weights = np.abs(fmin(cost,np.ones(6),xtol=1e-12,maxiter=1e5,ftol=1e-12))
print(c_weights)
```

```python
#Sanity check on controls

controls_gen = generative_controller_model(robot,path_controller,X_traj,sigQ=c_weights[0],sigR=c_weights[1],Kfb=c_weights[2:].reshape(2,2),lookahead=5, Nsteps=50, dt=0.15)
plt.plot(controls_gen)
plt.plot(controls_gt,'--')
plt.title(cost(c_weights))
plt.grid()
plt.show()
```
